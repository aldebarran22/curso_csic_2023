## Pruebas con matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Funciones sen, cos, exp

```python
x = np.arange(0,10,0.05)
```

```python
y = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(x)
```

```python
#fig = plt.figure()
plt.subplot2grid((2,2),(0,0))
plt.title('sin(x)')
plt.grid()
plt.plot(x, y, label='sin')
plt.legend()

plt.subplot2grid((2,2),(0,1))
plt.title('cos(x)')
plt.plot(x, y2, color='red')

plt.subplot2grid((2,2),(1,0), colspan=2)
plt.title('exp(x)')
plt.plot(x, y3, color='green')

# Ajustar la separación entre los subgráficos:
plt.tight_layout()
plt.show()
```

```python
plt.title('Funciones trigonométricas')
plt.grid(True, ls=':', color='0.75')
plt.plot(x, y, label='sin', color='red')
plt.plot(x, y2, label='cos', color='blue')
plt.legend();
```
