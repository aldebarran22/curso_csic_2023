```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = list(range(1,11))
y = [np.exp(i) for i in x]
y2 = [np.sqrt(i) for i in x]
```

```python
x
```

```python
y
```

```python
y2
```

```python
plt.plot(x,y);
```

```python
plt.plot(x,y2);
```

```python
# Subgrafico 1
plt.suptitle("Gráficas comparativas")
plt.subplot2grid((1,2),(0,0))
plt.title("Grafico 1")
plt.xlabel("x gráfica1")
plt.ylabel("y gráfica1")
plt.plot(x,y)

# Subgrafico 2
plt.subplot2grid((1,2),(0,1))
plt.title("Grafico 2")
plt.xlabel("x gráfica2")
plt.ylabel("y gráfica2")
plt.plot(x,y2)
plt.show()
```
